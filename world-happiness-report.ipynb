```python
# read and print data

import pandas as pd
data = pd.read_csv('2018.csv')
display(data)
```

```python
# mean, median, mode

x_mean = data["Score"].mean()
x_mode = data["Score"].mode()[0]
x_median = data["Score"].median()
data["Score"].fillna(x_mode, inplace = True)

print("Happiness Index Scores Analytics")
print("Mean:", round(x_mean, 3))
print("Mode:", round(x_mode, 3))
print("Median:", round(x_median, 3))
```

```python
# visualizations
import matplotlib.pyplot as plt
import seaborn as sns

# set style
sns.set(style="whitegrid")

# Happiness Score Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['Score'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.show()

# GDP per Capita vs Happiness Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GDP per capita', y='Score', data=data, hue='Score', palette='viridis')
plt.title('GDP per Capita vs Happiness Score')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.show()

# Social Support vs Happiness Score
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Social support', y='Score', data=data, hue='Score', palette='coolwarm')
plt.title('Social Support vs Happiness Score')
plt.xlabel('Social Support')
plt.ylabel('Happiness Score')
plt.show()

# Perceptions of Corruption across Countries
plt.figure(figsize=(12, 8))
corruption_sorted = data.sort_values('Perceptions of corruption', ascending=False)
sns.barplot(x='Perceptions of corruption', y='Country or region', data=corruption_sorted.head(10), palette='magma')
plt.title('Top 10 Countries with Highest Perception of Corruption')
plt.xlabel('Perceptions of Corruption')
plt.ylabel('Country or Region')
plt.show()

# Top 10 Happiest Countries
plt.figure(figsize=(12, 8))
happiest_countries = data.sort_values('Score', ascending=False)
sns.barplot(x='Score', y='Country or region', data=happiest_countries.head(10), palette='Blues_d')
plt.title('Top 10 Happiest Countries')
plt.xlabel('Happiness Score')
plt.ylabel('Country or Region')
plt.show()
```
